## **Plotting Sigmoid Function**

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Sigmoid

## $f_{w_1,w_2b}(x_1,x_2) = \frac{1}{1+e^{-(w_1x_1+w_2x_2+b)}}$

```python
## Define sigmoid function
def sigmoid_2d(x1,x2,w1,w2,b):
  return 1/(1+np.exp(-(w1*x1 + w2*x2 + b)))
```

```python
sigmoid_2d(0.2,0.4,0.3,0.5,1)  # sigmoid goes for 0 to 1
```

## 3d Plotting in matplotlib 

https://jakevdp.github.io/PythonDataScienceHandbook/04.12-three-dimensional-plotting.html

```python
from mpl_toolkits import mplot3d
```

```python
X1 = np.linspace(-10,10,100)
X2 = np.linspace(-10,10,80)
XX1, XX2 = np.meshgrid(X1,X2)
print(X1.shape, X2.shape, XX1.shape, XX2.shape)
```

```python
w1 =0.3
w2 = 0.3
b = 0

Y = sigmoid_2d(XX1,XX2, w1,w2,b)
```

```python
w1 =0.4 #@param {type: "slider", min: -2, max: 2,step:0.1}
w2 = 1 #@param {type: "slider", min: -2, max: 2,step:0.1}
b = 2.5 #@param {type: "slider", min: -10, max: 10,step:0.1}
Y = sigmoid_2d(XX1,XX2, w1,w2,b)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(XX1, XX2, Y, cmap='viridis')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y');
```
